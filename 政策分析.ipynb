```python
import os
import re
import jieba
from pyecharts import options as opts
from pyecharts.charts import WordCloud
from pyecharts.globals import SymbolType
path = 'data/政策数据/'
with open("data/stopwords.txt", "r", encoding="utf-8") as f:
    sw = [i.replace("\n", "") for i in f.readlines()]
words = []
sw.append("\n")
for i in os.listdir(path):
    fn = path+i
    txt = open(fn, 'r', encoding='utf-8').read()
    words += jieba.lcut(txt)
_ = {}
for i in words:
    if i not in sw and len(i) > 1 and not bool(re.search(r'\d', i)):
        _[i] = _.get(i, 0) + 1
word = list(_.items())
c = (
    WordCloud()
    .add("", word, word_size_range=[20, 100], shape=SymbolType.TRIANGLE)
    .set_global_opts(title_opts=opts.TitleOpts(title="疫情背景下一带一路相关政策词云图"))
)
c.render_notebook()
```
