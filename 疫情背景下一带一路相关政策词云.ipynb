## 疫情背景下一带一路相关政策词云图

### 数据来源
* 中华人民共和国中央人民政府 http://www.gov.cn/
* 新华社 http://xinhuanet.com/
* 人民网 http://www.people.com.cn/

### 数据处理与可视化
* 将所有相关政策文本爬取后保存在数据集中
* Python读取政策文件并添加到文本库
* 使用jieba包进行分词
* 去除停用词，并去除已知关键词“一带”与“一路”
* 统计词频
* 使用pyecharts绘制词云图

```python
import os
import re
import jieba
from pyecharts import options as opts
from pyecharts.charts import WordCloud
from pyecharts.globals import SymbolType
path = 'data/政策数据/'
with open("data/stopwords.txt", "r", encoding="utf-8") as f:
    sw = [i.replace("\n", "") for i in f.readlines()]
words = []
sw.append("\n")
sw += ['一带', '一路', 'http', 'Post', 'htm', 'cn', 'gov']
for i in os.listdir(path):
    fn = path+i
    txt = open(fn, 'r', encoding='utf-8').read()
    words += jieba.lcut(txt)
_ = {}
for i in words:
    if i not in sw and len(i) > 1 and not bool(re.search(r'\d', i)):
        _[i] = _.get(i, 0) + 1
word = list(_.items())
c = (
    WordCloud()
    .add("", word, word_size_range=[20, 100], shape=SymbolType.TRIANGLE)
    .set_global_opts(title_opts=opts.TitleOpts(title="疫情背景下一带一路相关政策词云图"))
)
c.render_notebook()
# c.render("renders/wordcloud.html")
```
